# Business Transformation and Modeling

```python
query = """
SELECT 
    ROW_NUMBER() OVER ( ORDER BY ci.customer_id ) AS customer_key,
    ci.customer_id,
    ci.customer_firstname,
    ci.customer_lastname
FROM silver.crm_customers ci
"""

df = spark.sql(query)
df.display()
```

# Write it to Gold Table


```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("gold.dim_customers")
)
```
